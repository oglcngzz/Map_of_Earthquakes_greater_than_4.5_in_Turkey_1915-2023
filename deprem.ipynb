```python
%pip install folium
import folium
import json
from urllib.request import urlopen
```

```python
m = folium.Map(location=[40.549835, 34.953724], zoom_start=6)

def get_color(buyukluk):
    if buyukluk > 7:
        return 'black'
    elif 6 <= buyukluk <= 7:
        return 'brown'
    elif 5 <= buyukluk < 6:
        return 'red'
    elif 4 <= buyukluk < 5:
        return 'orange'
    else:
        return None

def get_radius(buyukluk):
    return min(buyukluk * 2, 1)  
for index, row in df.iterrows():
    enlem = row['Enlem']
    boylam = row['Boylam']
    buyukluk = row['ML'] 
    
    if buyukluk >= 4.5:  
        color = get_color(buyukluk)
        
        if color:
            radius = get_radius(buyukluk)
            label = f"Büyüklük: {buyukluk}"
            folium.CircleMarker(
                location=[enlem, boylam],
                radius=radius,
                color=color,
                fill=True,
                fill_color=color,
                popup=label  # Popup'ta büyüklük bilgisi gösterilir
            ).add_to(m)

legend_html = '''
     <div style="position: fixed; 
     bottom: 50px; left: 50px; width: 150px; height: 130px; 
     border:2px solid grey; z-index:9999; font-size:14px;
     ">&nbsp; Deprem Büyüklüğü <br>
     &nbsp; 7+ : Siyah <br>
     &nbsp; 6-7 : Kahverengi <br>
     &nbsp; 5-6 : Kırmızı <br>
     &nbsp; 4-4.5 : Turuncu <br>
     </div>
     '''

m.get_root().html.add_child(folium.Element(legend_html))

```
